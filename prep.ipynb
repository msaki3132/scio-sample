```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/customers-2000000.csv")
df.head()
```

```python
dfs = np.array_split(df, 20)
```

```python
import csv 

for i, dfi in enumerate(dfs):
    path = f"./data/customers/customers-2000000_{i}.csv"
    
    dfi.to_csv(path, index=None, quotechar='"', quoting=csv.QUOTE_ALL)
    dfi.to_parquet(path.replace("csv", "parquet"))
```
